# **instead of label encoding we use this factorize**

```python
data['gender'],label=pd.factorize(data['gender'])
data['smoking_status'],label=pd.factorize(data['smoking_status'])
data['ever_married'],label=pd.factorize(data['ever_married'])
data['Residence_type'],label=pd.factorize(data['Residence_type'])
data['work_type'],label=pd.factorize(data['work_type'])
data
```

```python
data.corr()
```

```python
x = data.drop(columns=['stroke','id'], axis=1)
x = x.reset_index(drop=True)
y=data['stroke']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
model=RandomForestClassifier(random_state=42)
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
accuracy = accuracy_score(y_test, y_predict)
accuracy
```

```python
# Pickle the trained model to a file
with open('RFmodel.pkl', 'wb') as f:
    pickle.dump(model, f)

# Load the pickled model
with open('RFmodel.pkl', 'rb') as f:
    model = pickle.load(f)
```
